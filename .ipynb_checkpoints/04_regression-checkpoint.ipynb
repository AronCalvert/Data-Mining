# CS4168 Project

## Data set

We are using the dataset tracks2026.csv available on Brightspace. The dataset
contains audio features and metadata for music tracks across multiple genres.
This dataset is an excerpt of the Spotify Tracks Dataset available at:
https://huggingface.co/datasets/maharshipandya/spotify-tracks-dataset.  The link to Hugging Face is included here only for reference.

# Regression (Predicting Popularity Score)

 in addition to classification, you must perform regression modelling using the original column popularity as the target variable.

## Requirements:
* Use a separate copy of the dataset in which the original popularity column is
retained (do not binarise).
* Train a regression model that performs as well as possible.
* Explore more than one modelling approach and justify your final choice.
* Discuss whether predicting exact popularity is more difficult than predicting
high/low popularity.
